```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#reading the csv file 
df = pd.read_csv("C:\\Users\\shrey\\OneDrive\\Desktop\\Project_IPL2022\\OrangeCap.csv")
df
```

```python
#Variation of position Quinton de Kock on the orange cap table
date = df['Day'].unique()
date
```

```python
filter = df['Player']=='Quinton de Kock'
```

```python
pos_k = df['Pos'].where(filter)
pos_k = pos_k.dropna()
pos_k = pos_k.values
```

```python
plt.plot(date,pos_k,color='cyan')
plt.xlabel("Date")
plt.ylabel("Position")
plt.title("Variation of position of Quinton de Kock")
plt.show()
```
